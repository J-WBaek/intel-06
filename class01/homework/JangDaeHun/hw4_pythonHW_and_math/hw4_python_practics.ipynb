```python
items = [1, 2, 3, 4, 'Hello', 6.24]

for k in range(0,len(items)):
  print(items[k])
print('----------')
for item in items:
  print(item)
print('----------')
items = [[1,2], [3,4], [5,6]]
for item in items:
  print(item[0], item[1])
print('----------')
for item1, item2 in items:
  print(item1, item2)
print('----------')
info = {'A' : 1, 'B' : 2, 'C' : 3}
for key in info:
  print(key, info[key])
print('----------')
for key, value in info.items():
  print(key, value)
```

```python
import numpy as np
from numpy import linalg as LA
c = np.array([[1,2,3],
              [-1,1,4]])
print(LA.norm(c, axis=0))
print(LA.norm(c, axis=1))
print(LA.norm(c, ord=1, axis=1))
print(LA.norm(c, ord=2, axis=1))
```

```python
A = np.array([[1,4,2,0], [9,5,0,0], [4,0,2,4], [6,1,8,3]])
x = np.array([1,2,3,4])
b = np.array([0,0,0,0])
n = 4
for i in range(0,n):
  val = 0.0
  for j in range(0,n):
    val += A[i,j] * x[j]
  b[i] = val

print("calculate=",b)

b = np.dot(A,x)
print("dot=",b)

b = np.matmul(A,x)
print("matmul=",b)

b = A@x
print("A@x=",b)

b = A*x
print("A*x=",b)
```
