```python
n = int(input("정수 n을 입력하세요: "))
num = 1

for i in range(n):
    for j in range(n):
        print(f"{num:2d}", end=' ')
        num += 1
    print()
```

```python
import numpy as np
nn = int(input())
A = np.zeros((nn,nn))
for ii in range(nn):
  for jj in range(nn):
    A[ii,jj] = ii*nn + jj + 1
print(A)
```

```python
reshape_A = A.reshape(-1,)
print(reshape_A)
```
